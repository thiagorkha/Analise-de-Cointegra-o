```python
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

def analyze_stock_with_prophet(stock_ticker, forecast_horizon=15, test_size=0.2):
    """
    Analisa uma ação com Prophet, incluindo validação cruzada e métricas de desempenho.

    Parâmetros:
    - stock_ticker: Ticker da ação (ex: 'PETR4.SA' para Petrobras).
    - forecast_horizon: Número de dias úteis para previsão (padrão: 15).
    - test_size: Proporção dos dados para teste (padrão: 20%).
    """

    # Baixar dados (últimos 2 anos)
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=730)).strftime('%Y-%m-%d')

    try:
        data = yf.download(stock_ticker, start=start_date, end=end_date)['Close']
    except Exception as e:
        print(f"\nErro ao baixar dados para {stock_ticker}: {e}")
        return

    # Verificar se há dados suficientes
    if len(data) < 30:
        print(f"\nErro: Não há dados suficientes para {stock_ticker}.")
        return

    # Calcular retornos diários (1-dimensional)
    returns = data.pct_change().dropna().reset_index()
    returns.columns = ['ds', 'y']  # Renomear para o formato do Prophet

    # Verificar se há retornos válidos
    if len(returns) < 30:
        print(f"\nErro: Não há retornos suficientes para {stock_ticker}.")
        return

    # Dividir em treino e teste
    train_size = int(len(returns) * (1 - test_size))
    train = returns.iloc[:train_size]
    test = returns.iloc[train_size:]

    # Treinar modelo Prophet
    model = Prophet(interval_width=0.95)  # Intervalo de confiança de 95%
    model.fit(train[['ds', 'y']])  # Garantir que as colunas são 1-dimensional

    # Validação cruzada
    try:
        df_cv = cross_validation(model, initial=f'{train_size} days', period='30 days', horizon=f'{forecast_horizon} days')
        df_p = performance_metrics(df_cv)
        rmse = np.mean(df_p['rmse'])
        mae = np.mean(df_p['mae'])
        print(f"\nMédias - CV RMSE: {rmse:.4f}, CV MAE: {mae:.4f}")
    except Exception as e:
        print(f"\nErro na validação cruzada: {e}")
        return

    # Prever para o horizonte especificado
    future = model.make_future_dataframe(periods=forecast_horizon, freq='B')  # Dias úteis
    forecast = model.predict(future)

    # Filtrar previsões
    forecast_values = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(forecast_horizon)

    # Calcular retorno acumulado
    cumulative_return = np.prod(1 + forecast_values['yhat']) - 1
    lower_bound = np.prod(1 + forecast_values['yhat_lower']) - 1
    upper_bound = np.prod(1 + forecast_values['yhat_upper']) - 1

    # Resultados
    print(f"\n[Análise para {stock_ticker} - {datetime.today().strftime('%Y-%m-%d')}]")
    print(f"- Retorno Esperado ({forecast_horizon} dias): {cumulative_return*100:.2f}%")
    print(f"- Intervalo de Confiança: [{lower_bound*100:.2f}%, {upper_bound*100:.2f}%]")

    # Visualização
    plt.figure(figsize=(12, 6))
    model.plot(forecast, uncertainty=True)
    plt.title(f'Previsões para {stock_ticker} ({forecast_horizon} dias)')
    plt.xlabel('Data')
    plt.ylabel('Retorno Diário')
    plt.show()

# Exemplo de uso
analyze_stock_with_prophet(
    stock_ticker='VALE3.SA',  # Petrobras
    forecast_horizon=15,      # Previsão para 15 dias úteis
    test_size=0.2             # 20% dos dados para teste
)
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

def analyze_stock_with_prophet(stock_ticker, forecast_horizon=15, test_size=0.2):
    """
    Analisa uma ação com Prophet, incluindo validação cruzada e métricas de desempenho.

    Parâmetros:
    - stock_ticker: Ticker da ação (ex: 'PETR4.SA' para Petrobras).
    - forecast_horizon: Número de dias úteis para previsão (padrão: 15).
    - test_size: Proporção dos dados para teste (padrão: 20%).
    """

    # Baixar dados (últimos 5 anos para mais dados históricos)
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=5825)).strftime('%Y-%m-%d')  # 5 anos

    try:
        data = yf.download(stock_ticker, start=start_date, end=end_date)['Close']
    except Exception as e:
        print(f"\nErro ao baixar dados para {stock_ticker}: {e}")
        return

    # Verificar se há dados suficientes
    if len(data) < 30:
        print(f"\nErro: Não há dados suficientes para {stock_ticker}.")
        return

    # Calcular retornos diários (1-dimensional)
    returns = data.pct_change().dropna().reset_index()
    returns.columns = ['ds', 'y']  # Renomear para o formato do Prophet

    # Remover outliers (valores extremos)
    returns = returns[(returns['y'] >= returns['y'].quantile(0.01)) &
                      (returns['y'] <= returns['y'].quantile(0.99))]

    # Verificar se há retornos suficientes
    if len(returns) < 30:
        print(f"\nErro: Não há retornos suficientes para {stock_ticker}.")
        return

    # Dividir em treino e teste
    train_size = int(len(returns) * (1 - test_size))
    train = returns.iloc[:train_size]
    test = returns.iloc[train_size:]

    # Configurar o modelo Prophet
    model = Prophet(
        interval_width=0.90,  # Reduzir o intervalo de confiança para 90%
        changepoint_prior_scale=0.01,  # Menor sensibilidade a mudanças de tendência
        seasonality_prior_scale=0.1,   # Menor influência da sazonalidade
        yearly_seasonality=False,      # Desativar sazonalidade anual (se não for relevante)
        daily_seasonality=False,       # Desativar sazonalidade diária
        weekly_seasonality=True        # Manter sazonalidade semanal
    )

    # Adicionar feriados (exemplo para o Brasil)
    feriados = pd.DataFrame({
        'holiday': 'feriado',
        'ds': pd.to_datetime([
            '2024-01-01', '2024-02-12', '2024-03-29', '2024-04-21', '2024-05-01',
            '2024-09-07', '2024-10-12', '2024-11-02', '2024-11-15', '2024-12-25'
        ]),
        'lower_window': -1,  # Dias antes do feriado
        'upper_window': 1    # Dias após o feriado
    })
    model.add_country_holidays(country_name='BR')  # Feriados do Brasil
    model.holidays = feriados

    # Treinar o modelo
    model.fit(train[['ds', 'y']])

    # Validação cruzada
    try:
        df_cv = cross_validation(model, initial=f'{train_size} days', period='30 days', horizon=f'{forecast_horizon} days')
        df_p = performance_metrics(df_cv)
        rmse = np.mean(df_p['rmse'])
        mae = np.mean(df_p['mae'])
        print(f"\nMédias - CV RMSE: {rmse:.4f}, CV MAE: {mae:.4f}")
    except Exception as e:
        print(f"\nErro na validação cruzada: {e}")
        return

    # Prever para o horizonte especificado
    future = model.make_future_dataframe(periods=forecast_horizon, freq='B')  # Dias úteis
    forecast = model.predict(future)

    # Filtrar previsões
    forecast_values = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(forecast_horizon)

    # Calcular retorno acumulado
    cumulative_return = np.prod(1 + forecast_values['yhat']) - 1
    lower_bound = np.prod(1 + forecast_values['yhat_lower']) - 1
    upper_bound = np.prod(1 + forecast_values['yhat_upper']) - 1

    # Resultados
    print(f"\n[Análise para {stock_ticker} - {datetime.today().strftime('%Y-%m-%d')}]")
    print(f"- Retorno Esperado ({forecast_horizon} dias): {cumulative_return*100:.2f}%")
    print(f"- Intervalo de Confiança: [{lower_bound*100:.2f}%, {upper_bound*100:.2f}%]")

    # Visualização
    plt.figure(figsize=(12, 6))
    model.plot(forecast, uncertainty=True)
    plt.title(f'Previsões para {stock_ticker} ({forecast_horizon} dias)')
    plt.xlabel('Data')
    plt.ylabel('Retorno Diário')
    plt.show()

# Exemplo de uso
analyze_stock_with_prophet(
    stock_ticker='PETR4.SA',  # Petrobras
    forecast_horizon=10,      # Previsão para 15 dias úteis
    test_size=0.2             # 20% dos dados para teste
)
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

def analyze_stock_with_neural_network(stock_ticker, forecast_horizon=15, test_size=0.2):
    """
    Analisa uma ação com redes neurais, incluindo métricas de desempenho.

    Parâmetros:
    - stock_ticker: Ticker da ação (ex: 'PETR4.SA' para Petrobras).
    - forecast_horizon: Número de dias úteis para previsão (padrão: 15).
    - test_size: Proporção dos dados para teste (padrão: 20%).
    """

    # Baixar dados (últimos 5 anos para mais dados históricos)
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=5825)).strftime('%Y-%m-%d')  # 5 anos

    try:
        data = yf.download(stock_ticker, start=start_date, end=end_date)['Close']
    except Exception as e:
        print(f"\nErro ao baixar dados para {stock_ticker}: {e}")
        return

    # Verificar se há dados suficientes
    if len(data) < 30:
        print(f"\nErro: Não há dados suficientes para {stock_ticker}.")
        return

    # Calcular retornos diários
    returns = data.pct_change().dropna().values.reshape(-1, 1)

    # Criar features (usando retornos passados como entrada)
    X, y = [], []
    lookback = 10  # Usar os últimos 10 dias como entrada
    for i in range(lookback, len(returns) - forecast_horizon):
        X.append(returns[i - lookback:i].flatten())
        y.append(returns[i:i + forecast_horizon].mean())  # Média dos retornos no horizonte

    X = np.array(X)
    y = np.array(y)

    # Dividir em treino e teste
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)

    # Treinar modelo de rede neural
    model = MLPRegressor(
        hidden_layer_sizes=(100, 50),  # Arquitetura da rede
        max_iter=1000,                 # Número máximo de iterações
        random_state=42
    )
    model.fit(X_train, y_train)

    # Avaliar o modelo
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    print(f"\nMétricas de Desempenho - RMSE: {rmse:.4f}, MAE: {mae:.4f}")

    # Prever para o horizonte especificado
    last_data = returns[-lookback:].flatten().reshape(1, -1)  # Últimos dados disponíveis
    forecasted_returns = []
    for _ in range(forecast_horizon):
        predicted_return = model.predict(last_data)[0]
        forecasted_returns.append(predicted_return)
        # Atualizar os dados para a próxima previsão
        last_data = np.roll(last_data, -1)
        last_data[-1] = predicted_return

    # Calcular retorno acumulado
    cumulative_return = np.prod(1 + np.array(forecasted_returns)) - 1

    # Calcular intervalo de confiança (assumindo distribuição normal)
    mean_return = np.mean(forecasted_returns)
    std_return = np.std(forecasted_returns)
    lower_bound = mean_return - 1.96 * std_return
    upper_bound = mean_return + 1.96 * std_return

    # Resultados
    print(f"\n[Análise para {stock_ticker} - {datetime.today().strftime('%Y-%m-%d')}]")
    print(f"- Retorno Esperado ({forecast_horizon} dias): {cumulative_return*100:.2f}%")
    print(f"- Intervalo de Confiança: [{lower_bound*100:.2f}%, {upper_bound*100:.2f}%]")

    # Visualização
    plt.figure(figsize=(12, 6))
    plt.plot(range(forecast_horizon), forecasted_returns, label='Retornos Previstos', marker='o')
    plt.fill_between(range(forecast_horizon), lower_bound, upper_bound, color='gray', alpha=0.2, label='Intervalo de Confiança (95%)')
    plt.title(f'Previsões para {stock_ticker} ({forecast_horizon} dias)')
    plt.xlabel('Dias Futuros')
    plt.ylabel('Retorno Diário')
    plt.legend()
    plt.grid()
    plt.show()

# Exemplo de uso
analyze_stock_with_neural_network(
    stock_ticker='PETR4.SA',  # Petrobras
    forecast_horizon=15,      # Previsão para 10 dias úteis
    test_size=0.2             # 20% dos dados para teste
)
```

```python

!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from kerastuner.tuners import RandomSearch
import warnings

warnings.filterwarnings("ignore")

# Função para adicionar indicadores técnicos
def add_technical_indicators(data):
    # Média Móvel Simples (SMA)
    data['SMA_10'] = data['Close'].rolling(window=10).mean()
    data['SMA_50'] = data['Close'].rolling(window=50).mean()

    # Índice de Força Relativa (RSI)
    delta = data['Close'].diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
    rs = gain / loss
    data['RSI'] = 100 - (100 / (1 + rs))

    # Remover NaNs gerados pelos indicadores
    data.dropna(inplace=True)
    return data

# Função para criar o modelo LSTM com Keras Tuner
def build_lstm_model(hp, input_shape, forecast_horizon):
    model = Sequential()
    model.add(LSTM(
        units=hp.Int('units', min_value=32, max_value=256, step=32),
        return_sequences=True,
        input_shape=input_shape
    ))
    model.add(Dropout(hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)))
    model.add(LSTM(
        units=hp.Int('units', min_value=32, max_value=256, step=32),
        return_sequences=False
    ))
    model.add(Dropout(hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)))
    model.add(Dense(units=forecast_horizon))
    model.compile(optimizer='adam', loss='mse')
    return model

# Função principal
def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    """
    Analisa uma ação com LSTM, incluindo indicadores técnicos e tuning de hiperparâmetros.

    Parâmetros:
    - stock_ticker: Ticker da ação (ex: 'PETR4.SA' para Petrobras).
    - forecast_horizon: Número de dias úteis para previsão (padrão: 15).
    - test_size: Proporção dos dados para teste (padrão: 20%).
    """

    # Baixar dados (últimos 5 anos para mais dados históricos)
    end_date = datetime.today().strftime('%Y-%m-%d')
    start_date = (datetime.today() - timedelta(days=2825)).strftime('%Y-%m-%d')  # 5 anos

    try:
        data = yf.download(stock_ticker, start=start_date, end=end_date)
    except Exception as e:
        print(f"\nErro ao baixar dados para {stock_ticker}: {e}")
        return

    # Verificar se há dados suficientes
    if len(data) < 30:
        print(f"\nErro: Não há dados suficientes para {stock_ticker}.")
        return

    # Adicionar indicadores técnicos
    data = add_technical_indicators(data)

    # Selecionar colunas relevantes
    feature_columns = ['Close', 'Volume', 'SMA_10', 'SMA_50', 'RSI']
    data = data[feature_columns]

    # Normalizar os dados
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(data)

    # Criar features e targets
    X, y = [], []
    lookback = 30  # Usar os últimos 30 dias como entrada
    for i in range(lookback, len(scaled_data) - forecast_horizon):
        X.append(scaled_data[i - lookback:i])
        y.append(scaled_data[i:i + forecast_horizon, 0])  # Prever o preço de fechamento

    X = np.array(X)
    y = np.array(y)

    # Dividir em treino e teste
    split = int(len(X) * (1 - test_size))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    # Busca de hiperparâmetros com Keras Tuner
    tuner = RandomSearch(
        lambda hp: build_lstm_model(hp, input_shape=(X_train.shape[1], X_train.shape[2]), forecast_horizon=forecast_horizon),
        objective='val_loss',
        max_trials=10,
        executions_per_trial=1,
        directory='lstm_tuning',
        project_name='stock_prediction'
    )
    tuner.search(X_train, y_train, epochs=20, validation_data=(X_test, y_test), callbacks=[EarlyStopping(patience=3)])

    # Melhor modelo
    best_model = tuner.get_best_models(num_models=1)[0]

    # Treinar o melhor modelo
    history = best_model.fit(
        X_train, y_train,
        epochs=50,
        batch_size=32,
        validation_data=(X_test, y_test),
        callbacks=[EarlyStopping(patience=5)]
    )

    # Prever para o horizonte especificado
    last_data = scaled_data[-lookback:].reshape(1, lookback, -1)
    forecasted_values = best_model.predict(last_data)[0]

    # Inverter a normalização
    dummy_data = np.zeros((forecast_horizon, scaled_data.shape[1]))
    dummy_data[:, 0] = forecasted_values  # Apenas a coluna 'Close' é preenchida
    forecasted_values = scaler.inverse_transform(dummy_data)[:, 0]

    # Calcular métricas de desempenho
    y_pred = best_model.predict(X_test)
    dummy_data = np.zeros((len(y_pred), forecast_horizon, scaled_data.shape[1]))
    for i in range(len(y_pred)):
        dummy_data[i, :, 0] = y_pred[i]  # Apenas a coluna 'Close' é preenchida
    y_pred = scaler.inverse_transform(dummy_data.reshape(-1, scaled_data.shape[1]))[:, 0].reshape(len(y_pred), forecast_horizon)

    y_test_actual = np.zeros((len(y_test), forecast_horizon, scaled_data.shape[1]))
    for i in range(len(y_test)):
        y_test_actual[i, :, 0] = y_test[i]  # Apenas a coluna 'Close' é preenchida
    y_test_actual = scaler.inverse_transform(y_test_actual.reshape(-1, scaled_data.shape[1]))[:, 0].reshape(len(y_test), forecast_horizon)

    rmse = np.sqrt(mean_squared_error(y_test_actual.mean(axis=1), y_pred.mean(axis=1)))
    mae = mean_absolute_error(y_test_actual.mean(axis=1), y_pred.mean(axis=1))
    print(f"\nMétricas de Desempenho - RMSE: {rmse:.2f}, MAE: {mae:.2f}")

    # Resultados
    print(f"\n[Análise para {stock_ticker} - {datetime.today().strftime('%Y-%m-%d')}]")
    print(f"- Previsões para os próximos {forecast_horizon} dias:")
    for i, value in enumerate(forecasted_values):
        print(f"  Dia {i + 1}: {value:.2f}")

    # Visualização
    plt.figure(figsize=(12, 6))
    plt.plot(range(forecast_horizon), forecasted_values, label='Previsões', marker='o')
    plt.title(f'Previsões para {stock_ticker} ({forecast_horizon} dias)')
    plt.xlabel('Dias Futuros')
    plt.ylabel('Preço de Fechamento')
    plt.legend()
    plt.grid()
    plt.show()

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='BBAS3.SA',  # Petrobras
    forecast_horizon=3,      # Previsão para 10 dias úteis
    test_size=0.2             # 20% dos dados para teste
)
```

```python
!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, GRU, Input, concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
from kerastuner.tuners import RandomSearch
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings

warnings.filterwarnings("ignore")

def add_technical_indicators(data):
    try:
        # Verificação inicial da coluna Close
        if 'Close' not in data.columns:
            raise KeyError("Coluna 'Close' não encontrada nos dados originais")

        df = data.copy()

        # 1. Médias Móveis
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_50'] = df['Close'].rolling(window=50).mean()

        # 2. RSI
        delta = df['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        df['RSI'] = 100 - (100 / (1 + rs))

        # 3. ATR
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        df['TR'] = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # 4. MACD
        df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
        df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

        # 5. Bollinger Bands
        df['BB_MA20'] = df['Close'].rolling(window=20).mean()
        df['BB_STD'] = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_MA20'] + (df['BB_STD'] * 2)
        df['BB_Lower'] = df['BB_MA20'] - (df['BB_STD'] * 2)

        # 6. VWAP
        typical_price = (df['High'] + df['Low'] + df['Close']) / 3
        cumulative_typical = (typical_price * df['Volume']).cumsum()
        cumulative_volume = df['Volume'].cumsum()
        df['VWAP'] = cumulative_typical / cumulative_volume

        # 7. Diferenciação
        df['Close_diff'] = df['Close'].diff()

        # Remover colunas temporárias
        df.drop(['TR', 'EMA_12', 'EMA_26'], axis=1, inplace=True, errors='ignore')

        # Remover NaNs de forma segura
        initial_count = len(df)
        df.dropna(inplace=True)
        print(f"Removidos {initial_count - len(df)} registros com valores NaN")

        return df

    except Exception as e:
        print(f"Erro no processamento de indicadores: {str(e)}")
        return pd.DataFrame()

def build_lstm_model(hp, input_shape, forecast_horizon):
    input_layer = Input(shape=input_shape)

    # Camada LSTM Bidirecional
    lstm_layer = Bidirectional(LSTM(
        units=hp.Int('lstm_units', 64, 256, step=64),
        return_sequences=True
    ))(input_layer)
    lstm_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(lstm_layer)

    # Camada GRU
    gru_layer = GRU(
        units=hp.Int('gru_units', 64, 128, step=32),
        return_sequences=False
    )(lstm_layer)
    gru_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(gru_layer)

    # Camadas Densas
    dense_layer = Dense(
        units=hp.Int('dense_units', 32, 128, step=32),
        activation='relu'
    )(gru_layer)

    output_layer = Dense(forecast_horizon)(dense_layer)

    model = Model(inputs=input_layer, outputs=output_layer)

    optimizer = Adam(
        learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
    )

    model.compile(
        optimizer=optimizer,
        loss=Huber(),
        metrics=['mae']
    )
    return model

def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    try:
        # Download dos dados
        end_date = datetime.today().strftime('%Y-%m-%d')
        start_date = (datetime.today() - timedelta(days=5*365)).strftime('%Y-%m-%d')

        print(f"\n{'='*50}")
        print(f"Iniciando análise para {stock_ticker}")
        print(f"Período: {start_date} a {end_date}")

        data = yf.download(stock_ticker, start=start_date, end=end_date)

        if data.empty:
            print("Erro: Nenhum dado encontrado para o ticker informado")
            return

        print("\nDados brutos baixados:")
        print(f"- Colunas: {data.columns.tolist()}")
        print(f"- Registros: {len(data)}")
        print(f"- Período: {data.index[0].date()} a {data.index[-1].date()}")

        # Processamento de indicadores
        print("\nProcessando indicadores técnicos...")
        processed_data = add_technical_indicators(data)

        if processed_data.empty:
            print("Erro: Dados processados estão vazios após remoção de NaNs")
            return

        print("\nDados processados com sucesso:")
        print(f"- Colunas: {processed_data.columns.tolist()}")
        print(f"- Registros restantes: {len(processed_data)}")
        print(f"- Primeiros registros:\n{processed_data.head(3)}")

        # Verificação final da coluna Close
        if 'Close' not in processed_data.columns:
            print("Erro crítico: Coluna 'Close' não encontrada após processamento")
            return

        # Preparação de dados
        feature_columns = [col for col in processed_data.columns if col != 'Close']
        target_column = 'Close'

        # Normalização
        print("\nNormalizando dados...")
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(processed_data[feature_columns])
        scaled_target = processed_data[target_column].values.reshape(-1, 1)

        # Seleção de features
        print("Selecionando melhores preditores...")
        selector = SelectKBest(score_func=f_regression, k=min(10, len(feature_columns)))
        selected_features = selector.fit_transform(scaled_features, scaled_target)

        # Preparar sequências temporais
        print("\nPreparando dados para treinamento...")
        X, y = [], []
        lookback = 30

        for i in range(lookback, len(processed_data) - forecast_horizon):
            X.append(selected_features[i-lookback:i])
            y.append(scaled_target[i:i+forecast_horizon])

        X = np.array(X)
        y = np.array(y)

        print(f"\nFormato final dos dados:")
        print(f"- X shape: {X.shape}")
        print(f"- y shape: {y.shape}")

        if len(X) == 0:
            print("Erro: Dados insuficientes para treinamento após processamento")
            return

        # Divisão treino-teste temporal
        split = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Otimização de hiperparâmetros
        print("\nOtimizando hiperparâmetros...")
        tuner = RandomSearch(
            lambda hp: build_lstm_model(hp, (X_train.shape[1], X_train.shape[2]), forecast_horizon),
            objective='val_loss',
            max_trials=5,
            executions_per_trial=1,
            directory='stock_tuning',
            project_name='stock_forecast_v2'
        )

        tuner.search(
            X_train, y_train,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=[EarlyStopping(patience=3)],
            verbose=1
        )

        # Treinamento final
        print("\nTreinando modelo final...")
        best_model = tuner.get_best_models(num_models=1)[0]

        history = best_model.fit(
            X_train, y_train,
            epochs=50,
            batch_size=32,
            validation_data=(X_test, y_test),
            callbacks=[
                EarlyStopping(patience=5),
                ReduceLROnPlateau(factor=0.5, patience=3)
            ],
            verbose=1
        )

        # Previsão
        print("\nGerando previsões...")
        last_data = selected_features[-lookback:]
        forecast = best_model.predict(last_data.reshape(1, lookback, -1))[0]

        # Reverter normalização
        dummy = np.zeros((len(forecast), len(feature_columns)))
        dummy[:, selector.get_support(indices=True)] = forecast
        forecast = scaler.inverse_transform(dummy)[:, feature_columns.index('Close')]

        # Resultados
        print("\n=== RESULTADOS ===")
        print(f"Ativo analisado: {stock_ticker}")
        print(f"Último preço real: {processed_data['Close'].iloc[-1]:.2f}")
        print(f"Previsões para os próximos {forecast_horizon} dias:")
        for i, price in enumerate(forecast, 1):
            print(f"Dia {i}: {price:.2f} ({price - processed_data['Close'].iloc[-1]:.2f})")

        # Plot
        plt.figure(figsize=(12,6))
        plt.plot(forecast, marker='o', linestyle='--', color='#2ecc71')
        plt.title(f'Previsão de Preços - {stock_ticker}\n'
                 f"RMSE: {np.sqrt(mean_squared_error(y_test, best_model.predict(X_test))):.2f} | "
                 f"MAE: {mean_absolute_error(y_test, best_model.predict(X_test)):.2f}")
        plt.xlabel('Dias Futuros')
        plt.ylabel('Preço (R$)')
        plt.grid(alpha=0.3)
        plt.show()

    except Exception as e:
        print(f"\nErro durante a execução: {str(e)}")

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='BBAS3.SA',
    forecast_horizon=10
)
```

```python
!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, GRU, Input, concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
from kerastuner.tuners import RandomSearch
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings

warnings.filterwarnings("ignore")

def add_technical_indicators(data):
    try:
        # Verificação inicial da coluna Close
        if 'Close' not in data.columns:
            raise KeyError("Coluna 'Close' não encontrada nos dados originais")

        df = data.copy()

        # 1. Médias Móveis
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_50'] = df['Close'].rolling(window=50).mean()

        # 2. RSI
        delta = df['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        df['RSI'] = 100 - (100 / (1 + rs))

        # 3. ATR
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        df['TR'] = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # 4. MACD
        df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
        df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

        # 5. Bollinger Bands
        df['BB_MA20'] = df['Close'].rolling(window=20).mean()
        df['BB_STD'] = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_MA20'] + (df['BB_STD'] * 2)
        df['BB_Lower'] = df['BB_MA20'] - (df['BB_STD'] * 2)

        # 6. VWAP
        typical_price = (df['High'] + df['Low'] + df['Close']) / 3
        cumulative_typical = (typical_price * df['Volume']).cumsum()
        cumulative_volume = df['Volume'].cumsum()
        df['VWAP'] = cumulative_typical / cumulative_volume

        # 7. Diferenciação
        df['Close_diff'] = df['Close'].diff()

        # Remover colunas temporárias
        df.drop(['TR', 'EMA_12', 'EMA_26'], axis=1, inplace=True, errors='ignore')

        # Remover NaNs de forma segura
        initial_count = len(df)
        df.dropna(inplace=True)
        print(f"Removidos {initial_count - len(df)} registros com valores NaN")

        # Garantir que a coluna 'Close' ainda esteja presente
        if 'Close' not in df.columns:
            raise KeyError("Coluna 'Close' foi removida durante o processamento")
        return df
    except Exception as e:
        print(f"Erro no processamento de indicadores: {str(e)}")
        return pd.DataFrame()

def build_lstm_model(hp, input_shape, forecast_horizon):
    input_layer = Input(shape=input_shape)

    # Camada LSTM Bidirecional
    lstm_layer = Bidirectional(LSTM(
        units=hp.Int('lstm_units', 64, 256, step=64),
        return_sequences=True
    ))(input_layer)
    lstm_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(lstm_layer)

    # Camada GRU
    gru_layer = GRU(
        units=hp.Int('gru_units', 64, 128, step=32),
        return_sequences=False
    )(lstm_layer)
    gru_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(gru_layer)

    # Camadas Densas
    dense_layer = Dense(
        units=hp.Int('dense_units', 32, 128, step=32),
        activation='relu'
    )(gru_layer)
    output_layer = Dense(forecast_horizon)(dense_layer)

    model = Model(inputs=input_layer, outputs=output_layer)
    optimizer = Adam(
        learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
    )
    model.compile(
        optimizer=optimizer,
        loss=Huber(),
        metrics=['mae']
    )
    return model

def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    try:
        # Download dos dados
        end_date = datetime.today().strftime('%Y-%m-%d')
        start_date = (datetime.today() - timedelta(days=5 * 365)).strftime('%Y-%m-%d')
        print(f"\n{'=' * 50}")
        print(f"Iniciando análise para {stock_ticker}")
        print(f"Período: {start_date} a {end_date}")

        data = yf.download(stock_ticker, start=start_date, end=end_date)
        if data.empty or 'Close' not in data.columns:
            print(f"Erro: Nenhum dado válido encontrado para o ticker {stock_ticker}.")
            return

        print("\nDados brutos baixados:")
        print(f"- Colunas: {data.columns.tolist()}")
        print(f"- Registros: {len(data)}")
        print(f"- Período: {data.index[0].date()} a {data.index[-1].date()}")

        # Processamento de indicadores
        print("\nProcessando indicadores técnicos...")
        processed_data = add_technical_indicators(data)
        if processed_data.empty:
            print("Erro: Dados processados estão vazios após remoção de NaNs")
            return

        print("\nDados processados com sucesso:")
        print(f"- Colunas: {processed_data.columns.tolist()}")
        print(f"- Registros restantes: {len(processed_data)}")
        print(f"- Primeiros registros:\n{processed_data.head(3)}")

        # Verificação final da coluna Close
        if 'Close' not in processed_data.columns:
            print("Erro crítico: Coluna 'Close' não encontrada após processamento")
            return

        # Preparação de dados
        feature_columns = [col for col in processed_data.columns if col != 'Close']
        target_column = 'Close'

        # Normalização
        print("\nNormalizando dados...")
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(processed_data[feature_columns])
        target_scaler = StandardScaler()
        scaled_target = target_scaler.fit_transform(processed_data[target_column].values.reshape(-1, 1))

        # Seleção de features
        print("Selecionando melhores preditores...")
        selector = SelectKBest(score_func=f_regression, k=min(10, len(feature_columns)))
        selected_features = selector.fit_transform(scaled_features, scaled_target)

        # Preparar sequências temporais
        print("\nPreparando dados para treinamento...")
        X, y = [], []
        lookback = 30
        for i in range(lookback, len(processed_data) - forecast_horizon):
            X.append(selected_features[i - lookback:i])
            y.append(scaled_target[i:i + forecast_horizon])
        X = np.array(X)
        y = np.array(y)

        print(f"\nFormato final dos dados:")
        print(f"- X shape: {X.shape}")
        print(f"- y shape: {y.shape}")
        if len(X) == 0:
            print("Erro: Dados insuficientes para treinamento após processamento")
            return

        # Divisão treino-teste temporal
        split = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Otimização de hiperparâmetros
        print("\nOtimizando hiperparâmetros...")
        tuner = RandomSearch(
            lambda hp: build_lstm_model(hp, (X_train.shape[1], X_train.shape[2]), forecast_horizon),
            objective='val_loss',
            max_trials=5,
            executions_per_trial=1,
            directory='stock_tuning',
            project_name='stock_forecast_v2'
        )
        tuner.search(
            X_train, y_train,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=[EarlyStopping(patience=3)],
            verbose=1
        )

        # Treinamento final
        print("\nTreinando modelo final...")
        best_model = tuner.get_best_models(num_models=1)[0]
        history = best_model.fit(
            X_train, y_train,
            epochs=50,
            batch_size=32,
            validation_data=(X_test, y_test),
            callbacks=[
                EarlyStopping(patience=5),
                ReduceLROnPlateau(factor=0.5, patience=3)
            ],
            verbose=1
        )

        # Previsão
        print("\nGerando previsões...")
        last_data = selected_features[-lookback:]
        if last_data.shape != (lookback, selected_features.shape[1]):
            print("Erro: Formato inválido para os dados de entrada do modelo")
            return
        forecast = best_model.predict(last_data.reshape(1, lookback, -1))[0]

        # Reverter normalização
        forecast = target_scaler.inverse_transform(forecast.reshape(-1, 1)).flatten()

        # Resultados
        print("\n=== RESULTADOS ===")
        print(f"Ativo analisado: {stock_ticker}")
        print(f"Último preço real: {processed_data['Close'].iloc[-1]:.2f}")
        print(f"Previsões para os próximos {forecast_horizon} dias:")
        for i, price in enumerate(forecast, 1):
            print(f"Dia {i}: {price:.2f} ({price - processed_data['Close'].iloc[-1]:.2f})")

        # Plot
        plt.figure(figsize=(12, 6))
        plt.plot(forecast, marker='o', linestyle='--', color='#2ecc71')
        plt.title(f'Previsão de Preços - {stock_ticker}\n'
                  f"RMSE: {np.sqrt(mean_squared_error(y_test, best_model.predict(X_test))):.2f} | "
                  f"MAE: {mean_absolute_error(y_test, best_model.predict(X_test)):.2f}")
        plt.xlabel('Dias Futuros')
        plt.ylabel('Preço (R$)')
        plt.grid(alpha=0.3)
        plt.show()

    except Exception as e:
        print(f"\nErro durante a execução: {str(e)}")

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='BBAS3.SA',
    forecast_horizon=10
)
```

```python
!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, GRU, Input, concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
from kerastuner.tuners import RandomSearch
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings

warnings.filterwarnings("ignore")

def add_technical_indicators(data):
    try:
        # Verificação inicial da coluna Close
        if 'Close' not in data.columns:
            raise KeyError("Coluna 'Close' não encontrada nos dados originais")

        df = data.copy()

        # 1. Médias Móveis
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_50'] = df['Close'].rolling(window=50).mean()

        # 2. RSI
        delta = df['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        df['RSI'] = 100 - (100 / (1 + rs))

        # 3. ATR
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        df['TR'] = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # 4. MACD
        df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
        df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

        # 5. Bollinger Bands
        df['BB_MA20'] = df['Close'].rolling(window=20).mean()
        df['BB_STD'] = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_MA20'] + (df['BB_STD'] * 2)
        df['BB_Lower'] = df['BB_MA20'] - (df['BB_STD'] * 2)

        # 6. VWAP
        typical_price = (df['High'] + df['Low'] + df['Close']) / 3
        cumulative_typical = (typical_price * df['Volume']).cumsum()
        cumulative_volume = df['Volume'].cumsum()
        df['VWAP'] = cumulative_typical / cumulative_volume

        # 7. Diferenciação
        df['Close_diff'] = df['Close'].diff()

        # Remover colunas temporárias
        df.drop(['TR', 'EMA_12', 'EMA_26'], axis=1, inplace=True, errors='ignore')

        # Remover NaNs de forma segura
        initial_count = len(df)
        df.dropna(inplace=True)
        print(f"Removidos {initial_count - len(df)} registros com valores NaN")

        # Garantir que a coluna 'Close' ainda esteja presente
        if 'Close' not in df.columns:
            raise KeyError("Coluna 'Close' foi removida durante o processamento")
        return df
    except Exception as e:
        print(f"Erro no processamento de indicadores: {str(e)}")
        return pd.DataFrame()

def build_lstm_model(hp, input_shape, forecast_horizon):
    input_layer = Input(shape=input_shape)

    # Camada LSTM Bidirecional
    lstm_layer = Bidirectional(LSTM(
        units=hp.Int('lstm_units', 64, 256, step=64),
        return_sequences=True
    ))(input_layer)
    lstm_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(lstm_layer)

    # Camada GRU
    gru_layer = GRU(
        units=hp.Int('gru_units', 64, 128, step=32),
        return_sequences=False
    )(lstm_layer)
    gru_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(gru_layer)

    # Camadas Densas
    dense_layer = Dense(
        units=hp.Int('dense_units', 32, 128, step=32),
        activation='relu'
    )(gru_layer)
    output_layer = Dense(forecast_horizon)(dense_layer)

    model = Model(inputs=input_layer, outputs=output_layer)
    optimizer = Adam(
        learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
    )
    model.compile(
        optimizer=optimizer,
        loss=Huber(),
        metrics=['mae']
    )
    return model

def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    try:
        # Download dos dados
        end_date = datetime.today().strftime('%Y-%m-%d')
        start_date = (datetime.today() - timedelta(days=5 * 365)).strftime('%Y-%m-%d')
        print(f"\n{'=' * 50}")
        print(f"Iniciando análise para {stock_ticker}")
        print(f"Período: {start_date} a {end_date}")

        data = yf.download(stock_ticker, start=start_date, end=end_date)
        if data.empty or 'Close' not in data.columns:
            print(f"Erro: Nenhum dado válido encontrado para o ticker {stock_ticker}.")
            return

        print("\nDados brutos baixados:")
        print(f"- Colunas: {data.columns.tolist()}")
        print(f"- Registros: {len(data)}")
        print(f"- Período: {data.index[0].date()} a {data.index[-1].date()}")

        # Processamento de indicadores
        print("\nProcessando indicadores técnicos...")
        processed_data = add_technical_indicators(data)
        if processed_data.empty:
            print("Erro: Dados processados estão vazios após remoção de NaNs")
            return

        print("\nDados processados com sucesso:")
        print(f"- Colunas: {processed_data.columns.tolist()}")
        print(f"- Registros restantes: {len(processed_data)}")
        print(f"- Primeiros registros:\n{processed_data.head(3)}")

        # Verificação final da coluna Close
        if 'Close' not in processed_data.columns:
            print("Erro crítico: Coluna 'Close' não encontrada após processamento")
            return

        # Preparação de dados
        feature_columns = [col for col in processed_data.columns if col != 'Close']
        target_column = 'Close'

        # Normalização
        print("\nNormalizando dados...")
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(processed_data[feature_columns])
        target_scaler = StandardScaler()
        scaled_target = target_scaler.fit_transform(processed_data[target_column].values.reshape(-1, 1))

        # Seleção de features
        print("Selecionando melhores preditores...")
        selector = SelectKBest(score_func=f_regression, k=min(10, len(feature_columns)))
        selected_features = selector.fit_transform(scaled_features, scaled_target)

        # Preparar sequências temporais
        print("\nPreparando dados para treinamento...")
        X, y = [], []
        lookback = 30
        for i in range(lookback, len(processed_data) - forecast_horizon):
            X.append(selected_features[i - lookback:i])
            y.append(scaled_target[i:i + forecast_horizon])
        X = np.array(X)
        y = np.array(y)

        print(f"\nFormato final dos dados:")
        print(f"- X shape: {X.shape}")
        print(f"- y shape: {y.shape}")
        if len(X) == 0:
            print("Erro: Dados insuficientes para treinamento após processamento")
            return

        # Divisão treino-teste temporal
        split = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Otimização de hiperparâmetros
        print("\nOtimizando hiperparâmetros...")
        tuner = RandomSearch(
            lambda hp: build_lstm_model(hp, (X_train.shape[1], X_train.shape[2]), forecast_horizon),
            objective='val_loss',
            max_trials=5,
            executions_per_trial=1,
            directory='stock_tuning',
            project_name='stock_forecast_v2'
        )
        tuner.search(
            X_train, y_train,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=[EarlyStopping(patience=3)],
            verbose=1
        )

        # Treinamento final
        print("\nTreinando modelo final...")
        best_model = tuner.get_best_models(num_models=1)[0]
        history = best_model.fit(
            X_train, y_train,
            epochs=50,
            batch_size=32,
            validation_data=(X_test, y_test),
            callbacks=[
                EarlyStopping(patience=5),
                ReduceLROnPlateau(factor=0.5, patience=3)
            ],
            verbose=1
        )

        # Previsão
        print("\nGerando previsões...")
        last_data = selected_features[-lookback:]
        if last_data.shape != (lookback, selected_features.shape[1]):
            print("Erro: Formato inválido para os dados de entrada do modelo")
            return
        forecast = best_model.predict(last_data.reshape(1, lookback, -1))[0]

        # Reverter normalização
        forecast = target_scaler.inverse_transform(forecast.reshape(-1, 1)).flatten()

        # Garantir que os valores sejam escalares
        last_price = float(processed_data['Close'].iloc[-1])
        forecast = [float(price) for price in forecast]

        # Resultados
        print("\n=== RESULTADOS ===")
        print(f"Ativo analisado: {stock_ticker}")
        print(f"Último preço real: {last_price:.2f}")
        print(f"Previsões para os próximos {forecast_horizon} dias:")
        for i, price in enumerate(forecast, 1):
            print(f"Dia {i}: {price:.2f} ({price - last_price:.2f})")

        # Plot
        plt.figure(figsize=(12, 6))
        plt.plot(forecast, marker='o', linestyle='--', color='#2ecc71')
        plt.title(f'Previsão de Preços - {stock_ticker}\n'
                  f"RMSE: {np.sqrt(mean_squared_error(y_test.squeeze(), best_model.predict(X_test).squeeze())):.2f} | "
                  f"MAE: {mean_absolute_error(y_test.squeeze(), best_model.predict(X_test).squeeze()):.2f}")
        plt.xlabel('Dias Futuros')
        plt.ylabel('Preço (R$)')
        plt.grid(alpha=0.3)
        plt.show()

    except Exception as e:
        print(f"\nErro durante a execução: {str(e)}")

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='BBAS3.SA',
    forecast_horizon=10
)
```

```python
!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, GRU, Input, concatenate
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
from kerastuner.tuners import RandomSearch
from statsmodels.tsa.api import SimpleExpSmoothing
import warnings

warnings.filterwarnings("ignore")

def add_technical_indicators(data):
    try:
        # Verificação inicial da coluna Close
        if 'Close' not in data.columns:
            raise KeyError("Coluna 'Close' não encontrada nos dados originais")

        df = data.copy()

        # 1. Médias Móveis
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_50'] = df['Close'].rolling(window=50).mean()

        # 2. RSI
        delta = df['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        df['RSI'] = 100 - (100 / (1 + rs))

        # 3. ATR
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        df['TR'] = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # 4. MACD
        df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
        df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

        # 5. Bollinger Bands
        df['BB_MA20'] = df['Close'].rolling(window=20).mean()
        df['BB_STD'] = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_MA20'] + (df['BB_STD'] * 2)
        df['BB_Lower'] = df['BB_MA20'] - (df['BB_STD'] * 2)

        # 6. VWAP
        typical_price = (df['High'] + df['Low'] + df['Close']) / 3
        cumulative_typical = (typical_price * df['Volume']).cumsum()
        cumulative_volume = df['Volume'].cumsum()
        df['VWAP'] = cumulative_typical / cumulative_volume

        # 7. Diferenciação
        df['Close_diff'] = df['Close'].diff()

        # Remover colunas temporárias
        df.drop(['TR', 'EMA_12', 'EMA_26'], axis=1, inplace=True, errors='ignore')

        # Remover NaNs de forma segura
        initial_count = len(df)
        df.dropna(inplace=True)
        print(f"Removidos {initial_count - len(df)} registros com valores NaN")

        # Garantir que a coluna 'Close' ainda esteja presente
        if 'Close' not in df.columns:
            raise KeyError("Coluna 'Close' foi removida durante o processamento")
        return df
    except Exception as e:
        print(f"Erro no processamento de indicadores: {str(e)}")
        return pd.DataFrame()

def build_lstm_model(hp, input_shape, forecast_horizon):
    input_layer = Input(shape=input_shape)

    # Camada LSTM Bidirecional
    lstm_layer = Bidirectional(LSTM(
        units=hp.Int('lstm_units', 64, 256, step=64),
        return_sequences=True
    ))(input_layer)
    lstm_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(lstm_layer)

    # Camada GRU
    gru_layer = GRU(
        units=hp.Int('gru_units', 64, 128, step=32),
        return_sequences=False
    )(lstm_layer)
    gru_layer = Dropout(hp.Float('dropout', 0.2, 0.5, step=0.1))(gru_layer)

    # Camadas Densas
    dense_layer = Dense(
        units=hp.Int('dense_units', 32, 128, step=32),
        activation='relu'
    )(gru_layer)
    output_layer = Dense(forecast_horizon)(dense_layer)

    model = Model(inputs=input_layer, outputs=output_layer)
    optimizer = Adam(
        learning_rate=hp.Choice('learning_rate', [1e-2, 1e-3, 1e-4])
    )
    model.compile(
        optimizer=optimizer,
        loss=Huber(),
        metrics=['mae']
    )
    return model

def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    try:
        # Download dos dados
        end_date = datetime.today().strftime('%Y-%m-%d')
        start_date = (datetime.today() - timedelta(days=5 * 365)).strftime('%Y-%m-%d')
        print(f"\n{'=' * 50}")
        print(f"Iniciando análise para {stock_ticker}")
        print(f"Período: {start_date} a {end_date}")

        data = yf.download(stock_ticker, start=start_date, end=end_date)
        if data.empty or 'Close' not in data.columns:
            print(f"Erro: Nenhum dado válido encontrado para o ticker {stock_ticker}.")
            return

        print("\nDados brutos baixados:")
        print(f"- Colunas: {data.columns.tolist()}")
        print(f"- Registros: {len(data)}")
        print(f"- Período: {data.index[0].date()} a {data.index[-1].date()}")

        # Processamento de indicadores
        print("\nProcessando indicadores técnicos...")
        processed_data = add_technical_indicators(data)
        if processed_data.empty:
            print("Erro: Dados processados estão vazios após remoção de NaNs")
            return

        print("\nDados processados com sucesso:")
        print(f"- Colunas: {processed_data.columns.tolist()}")
        print(f"- Registros restantes: {len(processed_data)}")
        print(f"- Primeiros registros:\n{processed_data.head(3)}")

        # Verificação final da coluna Close
        if 'Close' not in processed_data.columns:
            print("Erro crítico: Coluna 'Close' não encontrada após processamento")
            return

        # Preparação de dados
        feature_columns = [col for col in processed_data.columns if col != 'Close']
        target_column = 'Close'

        # Normalização
        print("\nNormalizando dados...")
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(processed_data[feature_columns])
        target_scaler = StandardScaler()
        scaled_target = target_scaler.fit_transform(processed_data[target_column].values.reshape(-1, 1))

        # Seleção de features
        print("Selecionando melhores preditores...")
        selector = SelectKBest(score_func=f_regression, k=min(10, len(feature_columns)))
        selected_features = selector.fit_transform(scaled_features, scaled_target)

        # Preparar sequências temporais
        print("\nPreparando dados para treinamento...")
        X, y = [], []
        lookback = 30
        for i in range(lookback, len(processed_data) - forecast_horizon):
            X.append(selected_features[i - lookback:i])
            y.append(scaled_target[i:i + forecast_horizon])
        X = np.array(X)
        y = np.array(y)

        print(f"\nFormato final dos dados:")
        print(f"- X shape: {X.shape}")
        print(f"- y shape: {y.shape}")
        if len(X) == 0:
            print("Erro: Dados insuficientes para treinamento após processamento")
            return

        # Divisão treino-teste temporal
        split = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Otimização de hiperparâmetros
        print("\nOtimizando hiperparâmetros...")
        tuner = RandomSearch(
            lambda hp: build_lstm_model(hp, (X_train.shape[1], X_train.shape[2]), forecast_horizon),
            objective='val_loss',
            max_trials=5,
            executions_per_trial=1,
            directory='stock_tuning',
            project_name='stock_forecast_v2'
        )
        tuner.search(
            X_train, y_train,
            epochs=20,
            validation_data=(X_test, y_test),
            callbacks=[EarlyStopping(patience=3)],
            verbose=1
        )

        # Treinamento final
        print("\nTreinando modelo final...")
        best_model = tuner.get_best_models(num_models=1)[0]
        history = best_model.fit(
            X_train, y_train,
            epochs=50,
            batch_size=32,
            validation_data=(X_test, y_test),
            callbacks=[
                EarlyStopping(patience=5),
                ReduceLROnPlateau(factor=0.5, patience=3)
            ],
            verbose=1
        )

        # Previsão
        print("\nGerando previsões...")
        last_data = selected_features[-lookback:]
        if last_data.shape != (lookback, selected_features.shape[1]):
            print("Erro: Formato inválido para os dados de entrada do modelo")
            return
        forecast = best_model.predict(last_data.reshape(1, lookback, -1))[0]

        # Reverter normalização
        forecast = target_scaler.inverse_transform(forecast.reshape(-1, 1)).flatten()

        # Garantir que os valores sejam escalares
        last_price = float(processed_data['Close'].iloc[-1])
        forecast = [float(price) for price in forecast]

        # Cálculo de RMSE e MAE
        y_pred = best_model.predict(X_test)
        y_pred = target_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
        y_test_original = target_scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
        rmse = np.sqrt(mean_squared_error(y_test_original, y_pred))
        mae = mean_absolute_error(y_test_original, y_pred)

        # Cálculo do preço médio
        average_price = np.mean(y_test_original)

        # Normalização das métricas em porcentagem
        rmse_percentage = (rmse / average_price) * 100
        mae_percentage = (mae / average_price) * 100

        # Resultados
        print("\n=== RESULTADOS ===")
        print(f"Ativo analisado: {stock_ticker}")
        print(f"Último preço real: {last_price:.2f}")
        print(f"Previsões para os próximos {forecast_horizon} dias:")
        for i, price in enumerate(forecast, 1):
            print(f"Dia {i}: {price:.2f} ({price - last_price:.2f})")
        print(f"\nMétricas de Desempenho:")
        print(f"- RMSE (%): {rmse_percentage:.2f}%")
        print(f"- MAE (%): {mae_percentage:.2f}%")

        # Plot
        plt.figure(figsize=(12, 6))
        plt.plot(forecast, marker='o', linestyle='--', color='#2ecc71')
        plt.title(f'Previsão de Preços - {stock_ticker}\n'
                  f"RMSE (%): {rmse_percentage:.2f}% | MAE (%): {mae_percentage:.2f}%")
        plt.xlabel('Dias Futuros')
        plt.ylabel('Preço (R$)')
        plt.grid(alpha=0.3)
        plt.show()

    except Exception as e:
        print(f"\nErro durante a execução: {str(e)}")

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='BBAS3.SA',
    forecast_horizon=10
)
```

```python
!pip install keras-tuner
import yfinance as yf
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
from sklearn.preprocessing import RobustScaler
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Model
from tensorflow.keras.layers import LSTM, Dense, Dropout, Bidirectional, GRU, Input, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import Huber
from tensorflow.keras.regularizers import l1_l2
from kerastuner.tuners import RandomSearch
import warnings

warnings.filterwarnings("ignore")

def remove_outliers(df, column='Close'):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

def add_technical_indicators(data):
    try:
        if 'Close' not in data.columns:
            raise KeyError("Coluna 'Close' não encontrada nos dados originais")

        df = data.copy()
        # Médias Móveis
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_50'] = df['Close'].rolling(window=50).mean()

        # RSI
        delta = df['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss = -delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=14).mean()
        avg_loss = loss.rolling(window=14).mean()
        rs = avg_gain / avg_loss
        df['RSI'] = 100 - (100 / (1 + rs))

        # ATR
        high_low = df['High'] - df['Low']
        high_close = np.abs(df['High'] - df['Close'].shift())
        low_close = np.abs(df['Low'] - df['Close'].shift())
        df['TR'] = pd.concat([high_low, high_close, low_close], axis=1).max(axis=1)
        df['ATR_14'] = df['TR'].rolling(window=14).mean()

        # MACD
        df['EMA_12'] = df['Close'].ewm(span=12, adjust=False).mean()
        df['EMA_26'] = df['Close'].ewm(span=26, adjust=False).mean()
        df['MACD'] = df['EMA_12'] - df['EMA_26']
        df['MACD_Signal'] = df['MACD'].ewm(span=9, adjust=False).mean()

        # Bollinger Bands
        df['BB_MA20'] = df['Close'].rolling(window=20).mean()
        df['BB_STD'] = df['Close'].rolling(window=20).std()
        df['BB_Upper'] = df['BB_MA20'] + (df['BB_STD'] * 2)
        df['BB_Lower'] = df['BB_MA20'] - (df['BB_STD'] * 2)

        # VWAP
        typical_price = (df['High'] + df['Low'] + df['Close']) / 3
        cumulative_typical = (typical_price * df['Volume']).cumsum()
        cumulative_volume = df['Volume'].cumsum()
        df['VWAP'] = cumulative_typical / cumulative_volume

        # Diferenciação
        df['Close_diff'] = df['Close'].diff()

        # Remover colunas temporárias
        df.drop(['TR', 'EMA_12', 'EMA_26'], axis=1, inplace=True, errors='ignore')

        # Remover NaNs
        initial_count = len(df)
        df.dropna(inplace=True)
        print(f"Removidos {initial_count - len(df)} registros com valores NaN")
        return df
    except Exception as e:
        print(f"Erro no processamento de indicadores: {str(e)}")
        return pd.DataFrame()

def build_lstm_model(hp, input_shape, forecast_horizon):
    input_layer = Input(shape=input_shape)
    # Camada LSTM Bidirecional
    lstm_layer = Bidirectional(LSTM(
        units=hp.Int('lstm_units', 32, 256, step=32),
        return_sequences=True,
        kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)
    ))(input_layer)
    lstm_layer = BatchNormalization()(lstm_layer)
    lstm_layer = Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1))(lstm_layer)

    # Camada GRU
    gru_layer = GRU(
        units=hp.Int('gru_units', 32, 128, step=32),
        return_sequences=False,
        kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)
    )(lstm_layer)
    gru_layer = BatchNormalization()(gru_layer)
    gru_layer = Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1))(gru_layer)

    # Camadas Densas
    dense_layer = Dense(
        units=hp.Int('dense_units', 16, 128, step=16),
        activation='relu',
        kernel_regularizer=l1_l2(l1=1e-5, l2=1e-4)
    )(gru_layer)
    output_layer = Dense(forecast_horizon)(dense_layer)

    model = Model(inputs=input_layer, outputs=output_layer)
    optimizer = Adam(
        learning_rate=hp.Choice('learning_rate', [1e-4, 1e-3, 1e-2])
    )
    model.compile(
        optimizer=optimizer,
        loss=Huber(),
        metrics=['mae']
    )
    return model

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def analyze_stock_with_lstm(stock_ticker, forecast_horizon=15, test_size=0.2):
    try:
        # Download dos dados
        end_date = datetime.today().strftime('%Y-%m-%d')
        start_date = (datetime.today() - timedelta(days=10 * 365)).strftime('%Y-%m-%d')
        data = yf.download(stock_ticker, start=start_date, end=end_date)

        if data.empty or 'Close' not in data.columns:
            print(f"Erro: Nenhum dado válido encontrado para o ticker {stock_ticker}.")
            return

        # Remover outliers
        data = remove_outliers(data)

        # Processamento de indicadores
        processed_data = add_technical_indicators(data)
        if processed_data.empty:
            print("Erro: Dados processados estão vazios após remoção de NaNs")
            return

        # Preparação de dados
        feature_columns = [col for col in processed_data.columns if col != 'Close']
        target_column = 'Close'

        # Normalização robusta
        scaler = RobustScaler()
        scaled_features = scaler.fit_transform(processed_data[feature_columns])
        target_scaler = RobustScaler()
        scaled_target = target_scaler.fit_transform(processed_data[target_column].values.reshape(-1, 1))

        # Seleção de features com RFE
        selector = RFE(RandomForestRegressor(), n_features_to_select=min(10, len(feature_columns)))
        selected_features = selector.fit_transform(scaled_features, scaled_target)

        # Preparar sequências temporais
        X, y = [], []
        lookback = 30
        for i in range(lookback, len(processed_data) - forecast_horizon):
            X.append(selected_features[i - lookback:i])
            y.append(scaled_target[i:i + forecast_horizon])
        X, y = np.array(X), np.array(y)

        # Divisão treino-teste temporal
        split = int(len(X) * (1 - test_size))
        X_train, X_test = X[:split], X[split:]
        y_train, y_test = y[:split], y[split:]

        # Otimização de hiperparâmetros
        tuner = RandomSearch(
            lambda hp: build_lstm_model(hp, (X_train.shape[1], X_train.shape[2]), forecast_horizon),
            objective='val_loss',
            max_trials=10,
            executions_per_trial=1,
            directory='stock_tuning',
            project_name='stock_forecast_v2'
        )
        tuner.search(
            X_train, y_train,
            epochs=50,
            validation_data=(X_test, y_test),
            callbacks=[EarlyStopping(patience=10, restore_best_weights=True)],
            verbose=1
        )

        # Treinamento final
        best_model = tuner.get_best_models(num_models=1)[0]
        history = best_model.fit(
            X_train, y_train,
            epochs=100,
            batch_size=32,
            validation_data=(X_test, y_test),
            callbacks=[
                EarlyStopping(patience=10, restore_best_weights=True),
                ReduceLROnPlateau(factor=0.5, patience=5)
            ],
            verbose=1
        )

        # Previsão
        last_data = selected_features[-lookback:]
        forecast = best_model.predict(last_data.reshape(1, lookback, -1))[0]
        forecast = target_scaler.inverse_transform(forecast.reshape(-1, 1)).flatten()

        # Cálculo de métricas
        y_pred = best_model.predict(X_test)
        y_pred = target_scaler.inverse_transform(y_pred.reshape(-1, 1)).flatten()
        y_test_original = target_scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()
        rmse = np.sqrt(mean_squared_error(y_test_original, y_pred))
        mae = mean_absolute_error(y_test_original, y_pred)
        mape = mean_absolute_percentage_error(y_test_original, y_pred)

        # Resultados
        print("\n=== RESULTADOS ===")
        print(f"Ativo analisado: {stock_ticker}")
        last_price = processed_data['Close'].iloc[-1].item()  # Garante que é um escalar
        print(f"Último preço real: {last_price:.2f}")
        print(f"Previsões para os próximos {forecast_horizon} dias:")
        for i, price in enumerate(forecast, 1):
            print(f"Dia {i}: {price:.2f} ({price - last_price:.2f})")
        print(f"\nMétricas de Desempenho:")
        print(f"- RMSE: {rmse:.2f}")
        print(f"- MAE: {mae:.2f}")
        print(f"- MAPE (%): {mape:.2f}%")

        # Plot
        plt.figure(figsize=(12, 6))
        plt.plot(forecast, marker='o', linestyle='--', color='#2ecc71')
        plt.title(f'Previsão de Preços - {stock_ticker}\n'
                  f"RMSE: {rmse:.2f} | MAE: {mae:.2f} | MAPE (%): {mape:.2f}%")
        plt.xlabel('Dias Futuros')
        plt.ylabel('Preço (R$)')
        plt.grid(alpha=0.3)
        plt.show()
    except Exception as e:
        print(f"\nErro durante a execução: {str(e)}")

# Exemplo de uso
analyze_stock_with_lstm(
    stock_ticker='PETR4.SA',
    forecast_horizon=10
)
```
